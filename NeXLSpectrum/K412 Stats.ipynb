# K412 Fit Statistical Study
Examining various different fitting filters and least squares fitting algorithms using K412 data.
<div style="text-align: right"> NWMR 20-Nov-2019</div>

```julia
using Revise
using NeXLUncertainties
using NeXLCore
using NeXLSpectrum
using DataFrames
using Gadfly
# Configure output options
set_default_plot_size(10inch,4inch)
ENV["columns"]=160;
```

Load the unknown and reference spectra.  Summarize their metadata.

```julia
path = "K412 spectra/"
unks = map(i->readEMSA.("$(path)III-E K412[$(i)][4].msa"),0:4)
al2o3 = readEMSA("$(path)Al2O3 std.msa")
caf2 = readEMSA("$(path)CaF2 std.msa")
fe = readEMSA("$(path)Fe std.msa")
mgo = readEMSA("$(path)MgO std.msa")
sio2 = readEMSA("$(path)SiO2 std.msa");
asa(DataFrame,[unks...,al2o3,caf2,fe,mgo,sio2])
```

Create a detector model to match the spectra.

```julia
det = matching(fe, 132.0, 10)
```

Use the detector model to build a fitting filter.  We will examine 3 different fitting filters

  * <code>VariableWidthFilter</code> - Varies the width of the filter according to the detector FWHM at each channel
  * <code>ConstantWidthFilter</code> - Fixes the width of the filter at FWHM Mn K&alpha;
  * <code>GaussianFilter</code> - Variable width filter shaped like a vertically offset Gaussian
  
### Variable width filter

```julia
filt=buildfilter(VariableWidthFilter,det);
```

Build the filtered reference spectra which will be fit to the filtered unknown spectrum.

```julia
# Add characteristic peaks 
rr = [
    ( n"O", ktransitions, al2o3 ),
    ( n"Al", ktransitions, al2o3 ),
    ( n"Mg", ktransitions, mgo ),
    ( n"Si", ktransitions, sio2 ),
    ( n"Ca", alltransitions, caf2 ),
    ( n"Fe", alltransitions, fe )
]
rx = map(r ->(charFeature(r[1], r[2]), r[3]), rr)
refs = collect(Iterators.flatten(map(r -> filter(r[2], det, r[1], filt, 1.0/dose(r[2])),rx)))
```

Split the 5 unknowns into 100 subdivided spectra to increase the variance by a factor of 10.  Fit the 500 subdivided spectra using the default weighted linear least squares model and the <code>VariableWidthFilter</code>.

```julia
n = 4 # 100
res = mapreduce(unk->map(u->fit(u,filt,refs,false), subdivide(unk,n)),append!,unks);
@time res = mapreduce(unk->map(u->fit(u,filt,refs,false), subdivide(unk,n)),append!,unks);
```

Tabulate the 500 <code>FilterFitResults</code>.

```julia
df=asa(DataFrame, res, true)
```

Compute statistics on each column in the results table.

```julia
describe(df[:,2:end],:mean, :std, :min, :q25, :median, :q75, :max)
```

### Constant width filter
Repeat the weighted least squares fit using a <code>ConstantWidthFilter</code>.

```julia
filt=buildfilter(ConstantWidthFilter,det)
refs = collect(Iterators.flatten((r -> filter(r[2], det, r[1], filt, 1.0/dose(r[2]))).(rx)))
@time rescw = mapreduce(unk->map(u->fit(u,filt,refs,false), subdivide(unk,n)),append!,unks)
df = asa(DataFrame, rescw, true)
describe(df[:,2:end],:mean, :std, :min, :q25, :median, :q75, :max)
```

### Gaussian filter
Repeat the weighted least squares fit using a <code>GaussianFilter</code>.

```julia
filt=buildfilter(GaussianFilter,det)
refs = collect(Iterators.flatten((r -> filter(r[2], det, r[1], filt, 1.0/dose(r[2]))).(rx)))
@time resgf = mapreduce(unk->map(u->fit(u,filt,refs,false), subdivide(unk,n)),append!,unks)
df = asa(DataFrame, resgf, true)
describe(df[:,2:end],:mean, :std, :min, :q25, :median, :q75, :max)
```

### Pure ordinary least-squares

```julia
fito(unk::Spectrum, filt::TopHatFilter, refs::Vector{FilteredReference}, forcezeros=true) =
    filterfit(filter(unk, filt, 1.0/dose(unk)), refs, fitcontiguouso, forcezeros)

filt=buildfilter(GaussianFilter,det)
refs = collect(Iterators.flatten((r -> filter(r[2], det, r[1], filt, 1.0/dose(r[2]))).(rx)))
@time reso = mapreduce(unk->map(u->fito(u,filt,refs,false), subdivide(unk,n)),append!,unks)
df = asa(DataFrame, reso, true)
describe(df[:,2:end],:mean, :std, :min, :q25, :median, :q75, :max)
```

Study the variability between the fits using a <code>VariableWidthFilter</code> and various <code>A</code> and <code>B</code> widths.

```julia
dff=DataFrame(A=Float64[],B=Float64[],variable=Symbol[],mean=Float64[],std=Float64[])
for a in [ 0.8, 1.0, 1.2, 1.5, 2.0]
    for b in [ 1.5, 2.0, 2.5, 3.0, 4.0]
        println("a=$(a) b=$(b)")
        filt=buildfilter(VariableWidthFilter,det, a, b)
        refs = collect(Iterators.flatten((r -> filter(r[2], det, r[1], filt, 1.0/dose(r[2]))).(rx)))
        @time res = mapreduce(unk->map(u->fit(u,filt,refs,false), subdivide(unk,n)),append!,unks)
        df = asa(DataFrame, res)
        dff=vcat(dff, 
            hcat( 
                DataFrame(A=[a for _ in eachindex(refs)],B=[b for _ in eachindex(refs)]), 
                describe(df[:,2:end], :mean, :std)
            )
        )
    end
end
```
